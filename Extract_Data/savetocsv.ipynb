```python
import json
import pandas as pd
import numpy as np
```

```python
file = open('ferdowsi-data.txt', 'r')
count = 0
x =''
i = 0
while count < 16000:
    count += 1
    line = file.readline()
    if not line:
        break
    x += line
    if count %16 == 0:
        y = json.loads(x)
        print(y) 
        x = ''
        if i == 0 :
            df = pd.DataFrame(y["_source"], index=[i])
            i += 1
        else:
            df = df.append(pd.DataFrame(y["_source"], index=[i]))
            i +=1
    

file.close()

df
```

```python
df.to_csv('Data.csv',index=False)
```
