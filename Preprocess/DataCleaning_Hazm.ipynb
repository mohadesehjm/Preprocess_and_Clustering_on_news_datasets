# intoduction with hazm library 

First news articles normalize, then every words of context tokenize.

I find 10 frequent words in news articles.

then find 10 biagram(2-words) in news articles.


Remove stopwords in persian texts.

Again find 10 frequent words and biagram in news articles.

Finally I add one column that shows the number of Sentences in each news.

```python
from __future__ import unicode_literals
from hazm import *
import pandas as pd
```

```python
df = pd.read_csv("Data.csv")
df
```

```python
string = []
for i in range(10000):
    normalizer = Normalizer()
    if pd.isna(df['content'][i]) != True:
        normal = normalizer.normalize(df['content'][i])
        words = word_tokenize(normal)
        lemmatizer = Lemmatizer()
        for i in range(len(words)):
            words[i] = lemmatizer.lemmatize(words[i])

        string += words
        
len(string)
```

## save (and read) words

```python
# save
with open("listOfWords.txt", "w") as f:
    for s in string:
        f.write(str(s) +"\n")
        
        
# read     
st = []
with open("listOfWords.txt", "r") as f:
    for line in f:
        st.append(str(line.strip()))
```

```python
len(st)
```

# find 10 words of frequent

```python
from collections import Counter
      
Counter = Counter(string)

most_occur = Counter.most_common(10)
  
print(most_occur)
```

## biagram

```python
from nltk import ngrams

n = 2
bigrams = ngrams(string, n)
bigrams
```

```python
from collections import Counter
Counter = Counter(bigrams)

most_occur = Counter.most_common(10)
  
print(most_occur)
```

# remove stopwords

```python
import codecs
from hazm import Normalizer


nmz = Normalizer()
stops = "\n".join(
    sorted(
        list(
            set(
                [
                    nmz.normalize(w) for w in codecs.open('persian-stopwords-master/persian', encoding='utf-8').read().split('\n') if w
                ]
            )
        )
    )
)
stops
```

```python
normalizer = Normalizer()
string2 = []
for i in range(10000):
    normalizer = Normalizer()
    if pd.isna(df['content'][i]) != True:
        normal = normalizer.normalize(df['content'][i])
        words = word_tokenize(normal)
        lemmatizer = Lemmatizer()
        for i in range(len(words)):
            words[i] = lemmatizer.lemmatize(words[i])
        
        filtered_sentence = []
 
        for w in words:
            if w not in stops:
                filtered_sentence.append(w)

        string2 += filtered_sentence

# length of previous one was 4217353
len(string2)
```

# find 10 words of frequent without stop words

```python
from collections import Counter
  
Counter2 = Counter(string2)

most_occur2 = Counter2.most_common(10)
  
print(most_occur2)
```

# biagram

```python
from nltk import ngrams

n = 2
bigrams2 = ngrams(string2, n)
bigrams2
```

```python
from collections import Counter

Counter2 = Counter(bigrams2)
most_occur2 = Counter2.most_common(10)
  
print(most_occur2)
```

# add a new column that shows number of sentences

```python
for i in range(10000):
    if pd.isna(df['content'][i]) != True:
        normalizer = Normalizer()
        norm = normalizer.normalize(df['content'][i])
        df.loc[i,'numberOfSentences']= len(sent_tokenize(norm))
    else :
        df.loc[i,'numberOfSentences'] = 0
    
    
```
